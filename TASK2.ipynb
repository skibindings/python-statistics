```
all_data = pd.read_csv('s_anova_test.csv')
all_data.head(5)
```

```
def smart_anova(all_data):
  group_1=all_data[all_data['y']=='A']['x']
  group_2=all_data[all_data['y']=='B']['x']
  group_3=all_data[all_data['y']=='C']['x']

  stat, p_1=stats.shapiro(group_1)
  stat, p_2=stats.shapiro(group_2)
  stat, p_3=stats.shapiro(group_3)
  stat, p_bart=stats.bartlett(group_1,group_2,group_3)

  if p_1 >= 0.05 and p_2 >= 0.05 and p_3 >= 0.05 and p_bart >= 0.05:
    stat, p_anova = stats.f_oneway(group_1,group_2,group_3)
    return {"ANOVA": p_anova}
  else:
    stat, p_kw = stats.kruskal(group_1,group_2,group_3)
    return {"KW": p_kw}
```

```
print(smart_anova(all_data))
```
