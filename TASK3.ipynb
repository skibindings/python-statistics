```
all_data = pd.read_csv('test.csv')
all_data.head(5)
```

```
def smart_cor(all_data):
  group_1=all_data['V1']
  group_2=all_data['V3']

  stat, p_1=stats.shapiro(group_1)
  stat, p_2=stats.shapiro(group_2)

  if p_1 >= 0.05 and p_2 >= 0.05:
    stat, p = stats.pearsonr(group_1,group_2)
    return {"Pearson correlation coefficient": stat}
  else:
    stat, p = stats.spearmanr(group_1,group_2)
    return {"Spearman  correlation coefficient": stat}
```

```
print(smart_cor(all_data))
```
