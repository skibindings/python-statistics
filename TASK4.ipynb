```
all_data = pd.read_csv('test_data.csv')
all_data.head(5)
```

```
def most_significant(all_data):
  prep_data = all_data.apply(lambda x: x.value_counts(), axis=0)
  prep_data = prep_data.T
  prep_data

  prep_data['p-value'] = prep_data.apply(lambda x: stats.chisquare(x)[1], axis = 1)

  return prep_data.iloc[np.where(prep_data['p-value'] == prep_data['p-value'].min())[0]].index.tolist()
```

```
most_significant(all_data)
```
