```
all_data = pd.read_csv('step6.csv')
```

```
def filtered_cor(all_data):
  all_data = all_data.select_dtypes(exclude=['object'])
  cols = all_data.columns
  corr_matrix = all_data[cols].corr()
  np.fill_diagonal(corr_matrix.values, 0)
  return corr_matrix.abs().max().max()
```

```
print(filtered_cor(all_data))
```
