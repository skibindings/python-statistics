```
import matplotlib.pyplot as plt
import seaborn as sns

import numpy as np 
import pandas as pd 

from sklearn.linear_model import LinearRegression

from scipy import stats
```

```
def fill_na(all_data):
  train_data = all_data.dropna()
  test_data = all_data[all_data['y'].notna() == False]

  train_x = train_data.drop(columns='y')
  train_y = train_data['y']

  test_x = test_data.drop(columns='y')

  linreg = LinearRegression() 
  linreg.fit(train_x, train_y) 

  predicts=linreg.predict(test_x) 

  ret_data = all_data.copy()
  ret_data['y_full']=all_data['y']

  fill = pd.DataFrame(index=ret_data.index[ret_data.isnull().any(axis=1)], data=predicts,columns=['y'])
  ret_data = ret_data.fillna(fill)

  return ret_data
```
